```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the Excel file with optimization
# First, just read the sheet names to understand the structure
excel_file = "Voluntary-Registry-Offsets-Database--v2025-12-year-end.xlsx"
xls = pd.ExcelFile(excel_file)
print(f"Available sheets: {xls.sheet_names}")

# Load the data - adjust sheet name if needed
df = pd.read_excel(excel_file, sheet_name=0)

print(f"\nDataset shape: {df.shape}")
print(f"\nColumn names:")
print(df.columns.tolist())
print(f"\nFirst few rows:")
df.head()
```

```python
# Filter for Improved Forest Management and group by protocol
# Adjust column names based on your actual data structure
ifm_df = df[df['Category'].str.contains('Improved Forest Management|IFM', case=False, na=False)].copy()

print(f"Improved Forest Management records: {len(ifm_df)}")
print(f"\nUnique protocols: {ifm_df['Protocol'].unique()}")

# Group by protocol and sum the credits
credits_by_protocol = ifm_df.groupby('Protocol')['Vintage Credits'].sum().sort_values(ascending=False)

print(f"\nCredits by Protocol:")
print(credits_by_protocol)
```

```python
# Create visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Bar chart of credits by protocol
credits_by_protocol.plot(kind='barh', ax=ax1, color='steelblue')
ax1.set_xlabel('Total Credits Generated', fontsize=12)
ax1.set_ylabel('Protocol', fontsize=12)
ax1.set_title('Improved Forest Management Credits by Protocol', fontsize=14, fontweight='bold')
ax1.grid(axis='x', alpha=0.3)

# Pie chart showing distribution
colors = plt.cm.Set3(np.linspace(0, 1, len(credits_by_protocol)))
ax2.pie(credits_by_protocol.values, labels=credits_by_protocol.index, autopct='%1.1f%%', 
        colors=colors, startangle=90)
ax2.set_title('Distribution of IFM Credits by Protocol', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()

print(f"\nTotal Improved Forest Management Credits: {credits_by_protocol.sum():,.0f}")
```
